```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris(as_frame=True)
df = iris.frame
```

```python
import matplotlib.pyplot as plt
```

```python
# line bar 

# Example bar plot: petal_length vs petal_width
plt.bar(df['petal length (cm)'], df['petal width (cm)'])
plt.xlabel("species")
plt.ylabel("petal length")
plt.title('Ds asssignment no 3')
```

```python
# histogram

plt.hist(df)
plt.show()
```

```python
# pie chart
# Add species column (convert target numbers  names)
df['species'] = df['target'].map({0:"setosa", 1:"versicolor", 2:"virginica"})
# Count how many samples of each species
species_counts = df['species'].value_counts()
# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Iris Species (Pie Chart)")


plt.show()
```

```python
# scatter
plt.scatter(df['petal length (cm)'],df['petal width (cm)'],marker='o')
plt.show()
```

```python
# boxplot
# Create boxplot for sepal length grouped by species
df.boxplot(column="sepal length (cm)", by="species", grid=False, figsize=(6,4))
plt.title("Sepal Length Distribution by Species")
plt.suptitle("")   # removes the automatic 'Boxplot grouped by species' title
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```
